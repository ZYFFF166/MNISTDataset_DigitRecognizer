```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import random as rand
from pandas import DataFrame,Series
from sklearn.decomposition import PCA
```

```python
data = pd.read_csv('train.csv')
df1 = data.iloc[:,1:]
pca = PCA(n_components=2)
pca.fit(df1.T)
PCA(copy=True, n_components=2, whiten=False)
print(pca.explained_variance_ratio_)
df2 = DataFrame(pca.components_.T,columns=['PC-1','PC-2'])
df3 = pd.concat([df2,data["label"]],axis=1)
# print(data["label"])
# print(df3)
# ax = sb.scatterplot(x="PC-1", y="PC-2",hue="label",size =10,data=df3, palette="Set2")
```

```python
g = sb.FacetGrid(df3,hue = "label",height = 8,aspect=1.2).map(plt.scatter,"PC-1","PC-2",marker=".",edgecolor="w")
g.add_legend() 
```
